```haskell
f1 :: Double -> Double -> Double -> Double
-- f1 :: a -> a -> a -> a
f1 x y z = x ** (y/z)
```

```haskell
f2 :: Double -> Double -> Double -> Double
f2 x y z = sqrt (x/y - z)
```

```haskell
f3 :: Bool -> Bool -> [Bool]
f3 x y = [x == True] ++ [y]
```

```haskell
-- ++ is reserved for lists
-- f4 :: [a] -> [a] -> [a] -> Bool
f4 :: Eq a => [a] -> [a] -> [a] -> Bool
f4 x y z = x == (y ++ z)
```

```haskell
f1 1 2 3
f2 7 5 1
f3 True False
f4 [1] [2] [3]
```
