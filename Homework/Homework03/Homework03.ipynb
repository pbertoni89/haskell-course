### Question 1

Write a function that checks if the monthly consumption of an electrical device is bigger, equal, or smaller than the maximum allowed and
returns a message accordingly.

The function has to take the hourly consumption of an electrical device, the hours of daily use, and the maximum monthly consumption allowed.

(Monthly usage = consumption (kW) * hours of daily use (h) * 30 days).

```haskell
checkConsumption :: Float -> Float -> Float -> String
checkConsumption cons_kw_h usage_h max_cons_kw =
    let monthlyConsumption cons_kw_h usage_h = cons_kw_h * usage_h * 30.0
        tot = monthlyConsumption cons_kw_h usage_h
    in if tot > max_cons_kw then "Too much" else if tot < max_cons_kw then "Ok" else "Equal"
        -- | tot >  max_cons_kw = "Too much"
        -- | tot == max_cons_kw = "Equal"
        -- | otherwise          = "Ok"
```

```haskell
checkConsumption' :: Float -> Float -> Float -> String
checkConsumption' cons_kw_h usage_h max_cons_kw
    | tot > max_cons_kw = "Too much"
    | tot < max_cons_kw = "Ok"
    | otherwise         = "Equal"
    where
        monthlyConsumption cons_kw_h usage_h = cons_kw_h * usage_h * 30.0
        tot = monthlyConsumption cons_kw_h usage_h
```

```haskell
checkConsumption  0.2 2.3 13
checkConsumption  0.2 2.3 13.8
checkConsumption  0.2 2.3 14

checkConsumption' 0.2 2.3 13
checkConsumption' 0.2 2.3 13.8
checkConsumption' 0.2 2.3 14
```

### Question 2

Prelude:
  * We use the function `show :: a -> String` to transform any type into a String.
  * So `show 3` will produce `"3"` and `show (3 > 2)` will produce `"True"`.

In the previous function, return the excess/savings of consumption as part of the message.

```haskell
checkConsumption'' :: Float -> Float -> Float -> String
checkConsumption'' cons_kw_h usage_h max_cons_kw
    | delta > 0 = "Too much: exceeding " ++ show    delta  ++ " kW"
    | delta < 0 = "Ok: saving "          ++ show (- delta) ++ " kW"
    | otherwise = "Equal"      ++ show tot
    where
        monthlyConsumption cons_kw_h usage_h = cons_kw_h * usage_h * 30.0
        tot = monthlyConsumption cons_kw_h usage_h
        delta = tot - max_cons_kw
```

```haskell
checkConsumption'' 0.2 2.3 40
checkConsumption'' 0.2 2.3 0.2
```

### Question 4

Write a function that takes in two numbers and returns their quotient such that it is not greater than 1.

Return the number as a string, and in case the divisor is 0, return a message why the division is not possible.

To implement this function using both guards and if-then-else statements.  

```haskell
smallestQuotient :: Float -> Float -> Float
smallestQuotient x y
    | xfracy < 1.0 = xfracy
    | otherwise    = 1 / xfracy
    where
         xfracy = x / y

smallestQuotient 2 3
smallestQuotient 3 2
smallestQuotient 2 3 == smallestQuotient 3 2
```

```haskell
smallestQuotient' :: Float -> Float -> String
smallestQuotient' x y
    | y == 0      = "You cannot divide by zero"
    | otherwise   = show (smallestQuotient x y)

smallestQuotient' 2 3 == smallestQuotient' 3 2
smallestQuotient' 0 7
smallestQuotient' 7 0
```

### Question 5

Write a function that takes in two numbers and calculates the sum of square roots for the product and quotient of those numbers.

Write the function such that you use a where block inside a let expression and a let expression inside a where block. 

```haskell
sumSqProdQuot :: Float -> Float -> Float
sumSqProdQuot x y =
    sqrt xy_prod + sqrt xy_quot
    where
        xy_prod = x * y
        xy_quot = x / y

sumSqProdQuot' :: Float -> Float -> Float
sumSqProdQuot' x y =
    let
        xy_prod = x * y
        xy_quot = x / y
    in
        sqrt xy_prod + sqrt xy_quot


sumSqProdQuot 1 2
sumSqProdQuot' 1 2
```
