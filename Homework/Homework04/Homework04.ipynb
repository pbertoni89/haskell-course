### Question 1
Lets say you have the nested values defined bellow. How would you get the value of
4 by using only pattern matching in a function?

```haskell
nested :: [([Int], [Int])]
nested = [([1, 2], [3, 4]), ([5, 6], [7, 8])]
```

```haskell
takeThe1stTuple :: [([Int], [Int])] -> ([Int], [Int])
takeThe1stTuple (x : _)        = x
takeThe1stTuple nested

takeThe2ndListOf1stTuple :: [([Int], [Int])] -> [Int]
-- takeThe2ndListOf1stTuple (_, x : _)        = x
takeThe2ndListOf1stTuple    ((_, x) : _)      = x  -- parentheses are not redundant here
takeThe2ndListOf1stTuple nested
```

```haskell
takeTheFour :: [([Int], [Int])] -> Int
-- takeTheFour (([_], [_:x]), ([_], [_])) = x  -- avoid explanation
-- takeTheFour (([], [_:x]), ([], []))    = x  -- avoid explanation
-- takeTheFour ((_, [_:x]), _)            = x  -- avoid explanation
-- takeTheFour ((_, (_:x)) : _)           = x  -- avoid explanation
--
-- takeTheFour ((_, [_:x]) : _)           = x
-- Couldn't match expected type ‘Int’ with actual type ‘[a0]’
-- [_:x] means a single-element list where that element matches _:x — so x gets the tail [Int], not an Int.
-- You need to pattern match within the list directly.
--
-- takeTheFour ((_,  _:x) : _)            = x
-- takeTheFour ((_,  _:(_:x)) : _)        = x
-- Couldn't match expected type ‘Int’ with actual type ‘[Int]’
-- _:x splits into head + tail, so x captures the remaining list [4], not the element 4 itself.
-- You need a second : to go one level deeper.
takeTheFour ((_,  _:(x:_) ) : _)          = x
takeTheFour _                             = -1

takeTheFour nested
```

### Question 2
Write a function that takes a list of elements of any type and, if the list has 3 or more elements, it
removes them. Else, it does nothing. Do it two times, one with multiple function definitions and one with
case expressions.

```haskell
clearLongLists :: [a] -> [a]
clearLongLists l = case l of
    (x : y : []) -> [x, y]     -- why not [x, y]
    (x : [])     -> [x]        -- why not [x]
    (x : y : _ ) -> []
    []          ->  []

clearLongLists' :: [a] -> [a]
clearLongLists' [x, y]       = [x, y]
clearLongLists' [x]          = [x]
clearLongLists' (x : y : _ ) = []
clearLongLists' []           = []
```

```haskell
demoClearShortI :: [Int]
demoClearShortI = [1, 2]
demoClearLongI = [1, 2, 3]
:t demoClearShortI
clearLongLists demoClearShortI
clearLongLists demoClearLongI
clearLongLists' demoClearShortI
clearLongLists' demoClearLongI
```

```haskell
demoClearLongF  = [6.7, 7.8, 8.9]
:t demoClearLongF
clearLongLists demoClearLongF
```

### Question 4
Implement a function that returns True if a list is empty and False otherwise.

```haskell
isListEmpty :: [a] -> Bool
isListEmpty l = case l of
    [] -> True
    _  -> False
isListEmpty [2, 3]
isListEmpty []
```

###  Question 5
Write the implementation of the tail function using pattern matching. But, instead of failing if
the list is empty, return an empty list.

```haskell
myTail :: [a] -> [a]
myTail l = case l of
    []         -> []
    [x]        -> []
    [_, x]     -> [x]   -- exactly two elements
--   _         -> [:l]  -- error: A section must be enclosed in parentheses thus: (: l)
    (_ : x)    -> x

myTail []
myTail [1]
myTail [1, 2]
myTail [1, 2, 3]

-- tail []
tail [1]
tail [1, 2]
tail [1, 2, 3]

myTail [4, 5, 6] == tail [4, 5, 6]
```
